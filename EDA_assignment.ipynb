```python
import numpy as np

numerical_cols = df.select_dtypes(include=np.number).columns

outlier_counts = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_counts[col] = len(outliers)

display("Number of outliers per numerical column:")
display(outlier_counts)
```

**Reasoning**:
Based on the outlier counts, some columns have a significant number of outliers (e.g., FM, DP, ALTV, NSP). Removing these outliers might lead to a substantial loss of data. Capping or transformation might be more appropriate strategies. Given the varying number of outliers and the nature of the data (physiological measurements), capping the outliers to the upper and lower bounds determined by the IQR method seems like a reasonable approach to retain data while mitigating the impact of extreme values.



```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])

display("Outliers have been capped.")
```

## Visualize the data

### Subtask:
Create appropriate visualizations (e.g., histograms, box plots, scatter plots) to understand the distribution and relationships of variables.


**Reasoning**:
Create histograms, box plots, and scatter plots for selected numerical columns to visualize their distributions and relationships, and then display the plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols_to_viz = ['LB', 'AC', 'FM', 'UC', 'DL', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP']

# Create histograms
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(18, 16))
axes = axes.flatten()
for i, col in enumerate(numerical_cols_to_viz):
    sns.histplot(data=df, x=col, ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()

# Create box plots
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(18, 16))
axes = axes.flatten()
for i, col in enumerate(numerical_cols_to_viz):
    sns.boxplot(data=df, y=col, ax=axes[i])
    axes[i].set_title(f'Box Plot of {col}')
    axes[i].set_ylabel(col)
plt.tight_layout()
plt.show()

# Create scatter plots for relationships
scatter_pairs = [('LB', 'ASTV'), ('MSTV', 'MLTV'), ('Width', 'Tendency')]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))
axes = axes.flatten()
for i, (col1, col2) in enumerate(scatter_pairs):
    sns.scatterplot(data=df, x=col1, y=col2, ax=axes[i])
    axes[i].set_title(f'Scatter Plot of {col1} vs {col2}')
    axes[i].set_xlabel(col1)
    axes[i].set_ylabel(col2)
plt.tight_layout()
plt.show()
```
